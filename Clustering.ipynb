## Clustering

The goal of clustering is to find a natural grouping in data such that items in the same cluster are more similar to each other than those from diferent clusters. We will discuss one of the most popular clustering algorithms, Kmeans
which is widely used in both academia and industry. Clustering (or cluster analysis) is a technique that allows us to find groups of similar objects, objects that are more related to each other than to objects in other groups.

K-means is an iterative algorithm and it does

1. Cluster assignment step: assign point which is closer to cluster point

2. Move centroid step: move the centroid to the center of each cluster

```python
# K-means clustering

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


iris = datasets.load_iris()
X = iris.data
X = X[:,0:2]
y = iris.target

std = StandardScaler()
X_std = std.fit_transform(X)

cluster = KMeans(n_clusters=3, random_state=0, n_jobs=-1)

model = cluster.fit(X_std)

y_predict = model.predict(X_std)

#model.labels_
observation = [[0.8, 0.8]]
print("The observation's cluster is", model.predict(observation))


plt.scatter(X_std[:, 0], X_std[:, 1], c=y_predict, cmap="plasma")
plt.scatter(cluster.cluster_centers_[:, 0],   
            cluster.cluster_centers_[:, 1],
            marker='^', 
            c=[0, 1, 2], 
            s=100, 
            linewidth=2,
            cmap="plasma")
plt.scatter(0.8, 0.8, marker='*', linewidth=3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

Mini-batch KMeans is the same as KMeans. The only differnce is that KMeans is applied to randomly chosen "batch_size" number of observations. It reduces the computational cost significantly.  

```python
# Speeding up K-means clustering using minibatch

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans

iris = datasets.load_iris()
X = iris.data
X = X[:,0:2]
y = iris.target

std = StandardScaler()
X_std = std.fit_transform(X)

cluster = MiniBatchKMeans(n_clusters=3, random_state=0, batch_size=1000)

model = cluster.fit(X_std)

y_predict = model.predict(X_std)
#model.labels_
observation = [[0.8, 0.8]]
print("The observation's cluster is", model.predict(observation))


plt.scatter(X_std[:, 0], X_std[:, 1], c=y_predict, cmap="plasma")
plt.scatter(cluster.cluster_centers_[:, 0],   
            cluster.cluster_centers_[:, 1],
            marker='^', 
            c=[0, 1, 2], 
            s=100, 
            linewidth=4,
            cmap="plasma")

plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

The main disadvantage of clustering is that the number of cluster should be set prior the learning algorithm. TO overcome this problem, Meanshift can be used. 

```python
# Meanshift K-means 

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MeanShift

iris = datasets.load_iris()
X = iris.data
X = X[:,0:2]
y = iris.target

std = StandardScaler()
X_std = std.fit_transform(X)

cluster = MeanShift(n_jobs=-1)

model = cluster.fit(X_std)

y_predict = model.predict(X_std)
#model.labels_
observation = [[0.8, 0.8]]
print("The observation's cluster is", model.predict(observation))


plt.scatter(X_std[:, 0], X_std[:, 1], c=y_predict, cmap="plasma")
plt.scatter(0.8, 0.8, marker='*', linewidth=4, color='red')
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

Another clustering algorithm is DBSCAN. In this algorithm, we group observation into cluster with high density. 

```python
# DBSCAN clustering

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

iris = datasets.load_iris()
X = iris.data
X = X[:,0:2]
y = iris.target

std = StandardScaler()
X_std = std.fit_transform(X)

dbscan = DBSCAN()
clusters = dbscan.fit_predict(X_std)
# plot the cluster assignments
plt.scatter(X_std[:, 0], X_std[:, 1], c=clusters, cmap="plasma")
plt.scatter(0.8, 0.8, marker='*', linewidth=4, color='red')

plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
